## My Stuff

```python
locations = pd.read_csv('/Users/benjaminboughton/Documents/Projects/routing/SE_Asia_Locations.csv')

#Cleaning needed
locations['Long'] = pd.to_numeric(locations['Long'].astype(str))
locations['Lat'] = pd.to_numeric(locations['Lat'].astype(str))
locations.reset_index(drop=False, inplace=True)
locations.rename(columns={'index': 'icon_num'}, inplace=True)
locations['icon_num'] = locations['icon_num'] + 1
locations.loc[locations['png'] == 'None', 'png'] = None

#Mean long lat
long_mean = locations['Long'].mean()
lat_mean = locations['Lat'].mean()

#Tooltip
tooltip = ''

#Need list of lists of my long lats
longitudes = [long for long in locations['Long']]
latitudes = [lat for lat in locations['Lat']]

long_lats = [[longitudes[i], latitudes[i]] for i in range(0, len(longitudes))]
```

```python
#Initial map creation
map_plot_route = folium.Map(location = [long_mean, lat_mean], zoom_start = 5, tiles='CartoDB positron')

#Markers
locations.apply(lambda row: folium.Marker(location = [row['Long'], row['Lat']],
                                         popup=row['Name'],
                                         icon=plugins.BeautifyIcon(number = row['icon_num'],
                                                                  border_color='transparent',
                                                                  border_width =0.5,
                                                                  text_color='#A12830',
                                                                  inner_icon_style = 'margin:0px;')).add_to(map_plot_route), axis=1)


#Route
folium.PolyLine(long_lats).add_to(map_plot_route)

#Save as html
map_plot_route.save('map.html')
```

```python
#Images
for i in range(0, len(locations)):
    if locations['png'][i]:
        photoIcon = folium.features.CustomIcon(icon_image = '/Users/benjaminboughton/Documents/Projects/routing/pics/' + locations['png'][i],
                                       icon_size = (200,200))

        folium.Marker([locations['Long'][i]+0.1, locations['Lat'][i]], 
              popup=locations['Name'][i],
              tooltip = tooltip,
              icon=photoIcon).add_to(map_plot_route)
        
#Save as html
map_plot_route.save('map.html')
```
